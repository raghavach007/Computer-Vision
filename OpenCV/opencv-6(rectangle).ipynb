```python

import cv2
import numpy as np
import matplotlib.pyplot as plt
img = np.zeros((400, 400, 3), dtype=np.uint8)
cv2.rectangle(img, (50, 120), (300, 200), (243, 13, 47), 2)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('lalo.jpg')
cv2.rectangle(img, (100,100), (300,250), (0,0,255), -1)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.axis('off')
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
img = cv2.imread('lalo.jpg')
cv2.rectangle(img, (100,100), (400,300), (255,0,0), 3)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.axis('off')
plt.show()
```
